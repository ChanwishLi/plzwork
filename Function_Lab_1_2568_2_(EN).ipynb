**Q22**: Fahrenheit to Celsius function

Write a function $\texttt{FtoC}$ that takes a number for temperature in Fahrenheit and returns the temperature in Celsius. The formula is $C = (F-32)*5/9$.

**Input**:
* a real number for temperature in F.

**Output**:
* a real number for Celsius in two decimals

Example 1:
```
Enter temperature (F):89
31.67
```

Example 2:
```
Enter temperature (F):32
0.00
```

````python
# Implement FtoC function below
def FtoC(temp):
  C = (temp-32) * (5/9)
  return C


def main():
    temp = float(input("Enter temperature (F): "))
    print(f"{FtoC(temp):.2f}")
main()
````

**Q23**: Old or Young

Write a function `old_or_young(age)` that takes a number (age) and returns a message: `old` if the age is greater than 30, otherwise return `young`.

- An integer for age

**Output**:
- A string: `old` or `young`

```
old
```

**Q24** Rectangle

Write a function `rectangle(w,h)` that takes two numbers: Width `w` and Height `h`, and returns two values: the area and perimeter of the rectangle.


**Input**
- Real number, w
- Real number, h

**Output**
- Area, two decimal
- Perimeter, two decimal

**Requirement**
- You must create a `main()` function that receives input from the user, then calls the `rectangle(w,h)` function and stores the returned values in variables, and finally displays the output (see examples from previous questions).

**Example 1**

```
>Enter w: 5
>Enter h: 6
Area is 30.00
Perimeter is 22.00
```

**Example 2**

```
>Enter w: 10
>Enter h: 2.5
Area is 25.00
Perimeter is 25.00
```


````python
# implement rectangle function below
def rec(width, height):
  area = width * height
  perimiter = 2*width + 2*height
  return area, perimiter




# implement main function below
def main():
  width = float(input("Enter w: "))
  height = float(input("Enter h: "))
  area, perimeter = rec(width, height)
  print(f'Area is {area:.2f}')
  print(f'Perimeter is {perimeter:.2f}')

main()
````

**Q25**: Mean

Write a function `mean(numbers)` to calculate the mean of a list of numbers and return that mean value.


**Function Requirements**
- Argument: `numbers`: a list of integers
- Output: mean

Examples and function tests are provided in `main`.


````python
# implement mean function below
def mean(seq):
    total = 0
    for num in seq:
        total += num
    return total / len(seq)


def main():
    # Test 1
    seq1 = [1,2,3,4]
    result1 = mean(seq1)
    print(f'The seq is {seq1}. Your answer is {result1:.2f}. Expected answer is 2.50.')
    # Test 2
    seq2 = [4,4,4]
    result2 = mean(seq2)
    print(f'The seq is {seq2}. Your answer is {result2:.2f}. Expected answer is 4.00.')
    # Test 3
    seq3 = [-10, 30, 50, 70, 0, 1, 3]
    result3 = mean([-10, 30, 50, 70, 0, 1, 3])
    print(f'The seq is {seq3}. Your answer is {result3:.2f}. Expected answer is 20.57.')

main()
````
