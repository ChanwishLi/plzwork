**Q19**: Create a list from the user

Write a program to receive input from the user and create a list. The user will enter a value `n` first, then enter `n` more integers.

After that, find the mean of the list and create another list that subtracts the mean from every value.

**Extra**: Write two programs: 1) using `append`, 2) list comprehension

**Hint**: You can use `sum()`

Example 1:
```
Enter n: 3
Enter number: 1
Enter number: 1
Enter number: 1
The original list is [1, 1, 1]
The mean is 1.00
The new list is [0.0, 0.0, 0.0]
```

Example 2:

```
Enter n: 5
Enter number: 1
Enter number: 3
Enter number: 5
Enter number: 7
Enter number: 9
The original list is [1, 3, 5, 7, 9]
The mean is 5.00
The new list is [-4.0, -2.0, 0.0, 2.0, 4.0]
```


````python
# Using Append

n = int(input("Enter n: "))
lit = []
lit2 = []

for i in range(n):
  n_more = int(input("Enter number: "))
  lit.append(n_more)


mean = sum(lit) / len(lit)
for j in range(len(lit)):
  lit2.append(lit[j]-mean)

print(f"The original list is {lit}")
print(f"The mean is {mean:.2f}")
print(f"The new list is {lit2}")
````

**Q20**: FizzBuzz

Write a program that receives an integer $n$ and displays a list (`answer`) with the following characteristics:

* $\texttt{answer[i] == "FizzBuzz"}$ if $\texttt{i}$ is divisible by both 3 and 5
* $\texttt{answer[i] == "Fizz"}$ if $\texttt{i}$ is divisible by 3 only
* $\texttt{answer[i] == "Buzz"}$ if $\texttt{i}$ is divisible by 5 only
* $\texttt{answer[i] == i}$ if none of the above conditions are met


Example 1:

```
Enter n: 3
[1, 2, 'Fizz']
```

Example 2:

```
Enter n: 5
[1, 2, 'Fizz', 4, 'Buzz']
```

Example 3:

```
Enter n: 15
[1, 2, 'Fizz', 4, 'Buzz', 'Fizz', 7, 8, 'Fizz', 'Buzz', 11, 'Fizz', 13, 14, 'FizzBuzz']
```

````python
n = int(input("Enter n: "))
lit = []
for i in range(n):
  m = i + 1
  if m % 3 == 0 and m % 5 == 0:
    lit.append("FizzBuzz")
  elif m % 5 == 0:
    lit.append("Buzz")
  elif m % 3 == 0:
    lit.append("Fizz")
  else:
    lit.append(i+1)

print(lit)
````
